```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
bank=pd.read_csv('bank.csv',sep=";")
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='age',data=bank)
plt.show()
```

```python
sns.countplot(x='marital',data=bank)
```

```python
sns.countplot(x=bank[bank['loan']=="yes"]['loan'],hue='education',data=bank)
```

```python
sns.countplot(x=bank[bank['housing']=="yes"]['housing'],hue='marital',data=bank)
```

```python
sns.countplot(x=bank[bank['loan']=="yes"]['loan'],hue='marital',data=bank)
```

```python
data1=bank.groupby('education').apply(lambda x:(x[x['loan']=="yes"]['loan']).count())
data1.plot(kind='bar')
#print(data1)
plt.show()
```

```python
ax=sns.swarmplot(x="education",y="balance",hue="marital",data=bank)
```

```python
ax=sns.violinplot(x="balance",y="education",hue="marital",data=bank)
```

```python
ax=sns.stripplot(x="education",y="balance",hue="marital",data=bank)
```

```python
data2=bank.groupby('month').apply(lambda x:(x[x['loan']=="yes"]['loan']).count())
data2.plot(kind='bar')
#print(data2)
plt.show()
```

```python
data2=bank.groupby('month').apply(lambda x:(x[x['loan']=="no"]['loan']).count())
data2.plot(kind='bar')
#print(data2)
plt.show()
```
